## **Ejercicio 4**

Un cajero autom√°tico tiene 500.000 pesos disponibles.
Un usuario puede retirar varias veces, pero no puede retirar m√°s de lo que queda en el cajero. Escribe un programa que:

- Pregunte cu√°nto quiere retirar.

- Reste ese valor al saldo.

- Termine cuando el saldo llegue a 0 o el usuario escriba 0.

```python
### TU C√ìDIGO AQU√ç ###
saldo = 500000
while saldo > 0:
  retirar = float(input("¬øCu√°nto quiere retirar? "))
  if retirar == 0:
    print("operacion finalizada")
    break

  if retirar > saldo:
    print("Saldo insuficiente")
  else:
    saldo -= retirar
    print(f"Saldo restante: {saldo}")
print("sesion terminada")
```

## **Ejercicio 5**

Crea una funci√≥n raiz_cuadrada(num) que calcule la ra√≠z cuadrada de un n√∫mero.

Si el n√∫mero es negativo, devuelve: "Error: no se puede calcular la ra√≠z cuadrada de un n√∫mero negativo."

Si el valor no es num√©rico, devuelve: "Error: el valor debe ser num√©rico."

NOTA: utiliza import math y la funci√≥n math.sqrt(num). No olvides documentar la funci√≥n.

```python
### TU C√ìDIGO AQU√ç ###
import math
def raiz_cuadrada(num):
  """
  Calcula la ra√≠z cuadrada de un n√∫mero.

  Par√°metros:
  num (int o float): n√∫mero al que se le calcular√° la ra√≠z.

  Retorna:
  - La ra√≠z cuadrada si el n√∫mero es v√°lido.
  - Mensaje de error si el n√∫mero es negativo.
  - Mensaje de error si el valor no es num√©rico.
  """
  if not isinstance(num, (int, float)):
    return "Error: el valor debe ser num√©rico."

  if num < 0:
    return "Error: no se puede calcular la ra√≠z cuadrada de un n√∫mero negativo."
  return math.sqrt(num)
```

```python
print(raiz_cuadrada(16))
```

```python
print(raiz_cuadrada(-9))
```

```python
print(raiz_cuadrada("abc"))
```

## **Ejercicio 6**

Escribe una funci√≥n llamada es_par que reciba un n√∫mero entero como par√°metro y retorne True si el n√∫mero es par y False si es impar.

NOTA: % es el operador m√≥dulo en Python. Lo que hace es dividir un n√∫mero entre otro y devolver el residuo (lo que sobra de la divisi√≥n entera).

Si al dividir un n√∫mero en 2, el residuo es cero, entonces el n√∫mero es par, de lo contrario, es impar.

```python
### TU C√ìDIGO AQU√ç ###
def es_par(numero):
  """
  Determina si un n√∫mero es par o impar.
  Par√°metros:
  numero (int): N√∫mero entero a evaluar.
  Retorna: True si el n√∫mero es par, False si es impar.
  """
  if numero % 2 == 0:
    return True
  else:
    return False
```

```python
print(es_par(7))    # Deber√≠a imprimir False
```

## **Ejercicio 7**

Escribe un c√≥digo que permita al usuario crear una contrase√±a y que verifique que esta cumpla todas las siguientes condiciones:

- Tiene al menos 8 caracteres.

- Contiene al menos una letra may√∫scula.

- Contiene al menos una letra min√∫scula.

- Contiene al menos un n√∫mero.

Si cumple todas las condiciones, muestra "Contrase√±a v√°lida", de lo contrario, explica qu√© condici√≥n no se cumple.

```python
### TU C√ìDIGO AQU√ç ###
def validar_contrasena(contrasena):
  """
  Valida si una contrase√±a cumple con ciertos requisitos.
  - Tiene al menos 8 caracteres.
  - Contiene al menos una letra may√∫scula.
  - Contiene al menos una letra min√∫scula.
  - Contiene al menos un n√∫mero.
  """
  if len(contrasena) < 8:
    return "La contrase√±a debe tener al menos 8 caracteres."

  if not any(char.isupper() for char in contrasena):
    return "La contrase√±a debe contener al menos una letra may√∫scula."

  if not any(char.islower() for char in contrasena):
    return "La contrase√±a debe contener al menos una letra minuscula"

  if  not any(char.isdigit() for char in contrasena):
    return "La contrase√±a debe contener al menos un n√∫mero."

  return "Contrase√±a v√°lida"

clave = input("Ingrese una contrase√±a: ")
print(validar_contrasena(clave))
```

## **Ejercicio 8**

Crea una funci√≥n llamada "extraer_dominio" que extraiga el dominio de una p√°gina web a partir de una cadena que contenga un email de la siguiente forma:

usuario@dominio.com
    
Por ejemplo: si se le pasa "lmcastroco@gmail.com", la funci√≥n retornar√°: gmail.com

```python
### TU C√ìDIGO AQU√ç ###
def extraer_dominio(correo):
  """
  Extrae el dominio de una direcci√≥n de correo electr√≥nico.
  Par√°metros:
  correo (str): direcci√≥n de correo electr√≥nico.

  Retorna:
  El dominio del correo (parte despu√©s del @).
  """
  if "@" not in correo:
    return "Direcci√≥n de correo electr√≥nico inv√°lida"
  return correo.split("@")[1]
```

## **Ejercicio 10**

Vas conduciendo un carro muy r√°pido y eres detenido por un agente de tr√°nsito.

Escribe una funci√≥n "multa" que retorne uno de los 3 siguientes resultados:  

- Si la velocidad es 60 Km/h o menos, el resultado es "Sin multa".
- Si la velocidad est√° entre 60 y 80 Km/h (inclusive), el resultado es "Multa peque√±a".
- Si la velocidad es mayor a 80 Km/h, el resultado es "Multa grande".

Adem√°s, considera que si esto ocurre el d√≠a de tu cumplea√±os (codificado como par√°metro booleano que recibe la funci√≥n), tu velocidad podr√≠a ser 5 Km/ mayor en todos los casos.

```python
### TU C√ìDIGO AQU√ç ###
def multa(velocidad, es_cumplea√±os):
  """
  Determina el tipo de multa seg√∫n la velocidad del conductor.

  Si es el d√≠a de cumplea√±os, se permite un margen adicional de 5 km/h.

  Par√°metros:
  velocidad (int o float): Velocidad del veh√≠culo en km/h.
  cumpleanos (bool): Indica si es el d√≠a de cumplea√±os del conductor.

  Retorna:
  str: Tipo de multa correspondiente.
  """
  if es_cumplea√±os:
    velocidad -= 5
  if velocidad <= 60:
    return "Sin multa"
  elif 60 < velocidad <= 80:
    return "Multa peque√±a"
  else:
    return "Multa grande"
```

```python
multa(81, True)
```

```python
multa(81, False)
```

## **Ejercicio 15**

Un estudiante invierte 1.000 d√≥lares en un fondo que crece un 5% mensual. Escribe un programa que use while para calcular en cu√°ntos meses su inversi√≥n supera los 2.000 d√≥lares.

```python
### TU C√ìDIGO AQU√ç ###
inversion = 1000
tasa_interes = 0.05
meses = 0
while inversion < 2000:
  inversion *= 1 + tasa_interes
  meses += 1
print(f"La inversi√≥n supera los 2000 d√≥lares en {meses} meses.")
```

## **Ejercicio 16**

Una empresa quiere vender al menos 100 unidades de un producto. Cada d√≠a vende un n√∫mero aleatorio de unidades entre 1 y 10 (pista: usa import random y random.randint(1,10)). Escribe un programa con while que muestre:

- El total vendido cada d√≠a.

- En cu√°ntos d√≠as se logr√≥ la meta.

```python
### TU C√ìDIGO AQU√ç ###
import random
total_vendido = 0
dias = 0
while total_vendido < 100:
  unidades_vendidas = random.randint(1, 10)
  total_vendido += unidades_vendidas
  dias += 1
  print(f"Total vendido en el d√≠a {dias}: {total_vendido} unidades.")

print(f"\nMeta alcanzada en {dias} d√≠as con un total de {total_vendido} unidades vendidas.")
```
