##**Punto 3 Alemania**

```python
#Indicador = Código Banco Mundial
#PIB per cápita = NY.GDP.PCAP.CD
#Crecimiento del PIB	= NY.GDP.MKTP.KD.ZG
#Inflación	= FP.CPI.TOTL.ZG
#Desempleo = SL.UEM.TOTL.ZS
#Población total= SP.POP.TOTL
indicadores = {"NY.GDP.PCAP.CD": "pib_per_capita",
"NY.GDP.MKTP.KD.ZG": "crecimiento_pib",
"FP.CPI.TOTL.ZG": "inflacion",
"SL.UEM.TOTL.ZS": "desempleo",
"SP.POP.TOTL": "poblacion"}
```

```python
pais = ["DEU"] #Codigo ISO del pais que usa el Banco Mundial para identificar países en su base de datos, alemania es "DEU"
inicio3 = datetime.datetime(2010, 1, 1)
fin3 = datetime.datetime(2023, 12, 31)
```

```python
data_alemania = wbdata.get_dataframe(indicadores, country=pais, date=(inicio3, fin3))
```

```python
df_alemania = pd.DataFrame(data_alemania)
df_alemania.reset_index(inplace=True)
```
