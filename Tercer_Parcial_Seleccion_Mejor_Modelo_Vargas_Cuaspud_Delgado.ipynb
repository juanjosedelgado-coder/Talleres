```python
x = df.drop('Sales', axis=1)
y = df['Sales']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
numerical_features = x.select_dtypes(include=["int64","float64"]).columns
categorical_features = x.select_dtypes(include="object").columns

print(f"Columnas Numéricas: {list(numerical_features)}")
print(f"Columnas Categóricas: {list(categorical_features)}")
```

```python
# Crear el transformador para las categóricas
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
# Crear el preprocesador (ColumnTransformer)
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features)
    ], remainder= 'passthrough')
```

```python
# Ajustamos (fit) y transformamos el set de entrenamiento

x_train_processed = preprocessor.fit_transform(x_train)

# SÓLO transformamos el set de prueba (ya "aprendió" del train)

x_test_processed = preprocessor.transform(x_test)

print(f"\nForma de X_train procesado: {x_train_processed.shape}")
print(f"Forma de X_test procesado: {x_test_processed.shape}")
```

```python
# Inicializa el modelo, usa DecisionTreeRegressor
# Coloca como semilla random_state=42

tree_model = DecisionTreeRegressor(random_state=42)

# Entrena el modelo

tree_model.fit(x_train_processed, y_train)

# Realiza predicciones en el conjunto de prueba

y_pred_tree = tree_model.predict(x_test_processed)
```

```python
# Definimos la grilla de hiperparámetros
param_grid_tree = {
    'max_depth': [3, 5, 10, None],
    'min_samples_leaf': [1, 2, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# Creamos el modelo base
tree_model_grid = DecisionTreeRegressor(random_state=42)

# Configuramos GridSearchCV
grid_tree = GridSearchCV(
    estimator=tree_model_grid,
    param_grid=param_grid_tree,
    cv=5,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1
)

# Entrenamos
grid_tree.fit(x_train_processed, y_train)
```

```python
print("Mejores hiperparámetros:")
print(grid_tree.best_params_)
```

```python
print("Mejor RMSE:")
print(-grid_tree.best_score_)
```

```python
-grid_tree.best_score_
```

```python
best_tree_model = grid_tree.best_estimator_
```

```python
y_pred_best_tree = best_tree_model.predict(x_test_processed)
```

```python
rmse_tree = np.sqrt(mean_squared_error(y_test, y_pred_best_tree))

print("RMSE final:", rmse_tree)
```

```python
# Modelo Random Forest

rf_model = RandomForestRegressor(random_state=42)

# Hiperparámetros
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_leaf': [1, 2, 5]
}

# GridSearchCV
grid_rf = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid_rf,
    cv=5,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1
)
```

```python
# Entrenar
grid_rf.fit(x_train_processed, y_train)
```

```python
# Mejor modelo
best_rf = grid_rf.best_estimator_

# Predicciones
y_pred_rf = best_rf.predict(x_test_processed)
```

```python
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))

print("RMSE Random Forest:", rmse_rf)
```

```python
print("RMSE Árbol:", rmse_tree)
print("RMSE Random Forest:", rmse_rf)
```

##Gradient boosting

```python
gb_model = GradientBoostingRegressor(random_state=42)
```

```python
param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5]
}
```

```python
grid_gb = GridSearchCV(
    estimator=gb_model,
    param_grid=param_grid_gb,
    cv=3,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1
)
```

```python
grid_gb.fit(x_train_processed, y_train)
```

```python
print("Mejores hiperparámetros:")
print(grid_gb.best_params_)
```

```python
print("Mejor RMSE CV:")
print(-grid_gb.best_score_)
```

```python
best_gb = grid_gb.best_estimator_
```

```python
y_pred_gb = best_gb.predict(x_test_processed)
```

```python
rmse_gb = np.sqrt(mean_squared_error(y_test, y_pred_gb))

print("RMSE Gradient Boosting:", rmse_gb)
```

```python
r2_gb = r2_score(y_test, y_pred_gb)

print("R² Gradient Boosting:", r2_gb)
```

```python
print("RMSE Árbol:", rmse_tree)
print("RMSE Random Forest:", rmse_rf)
print("RMSE Gradient Boosting:", rmse_gb)
```

```python
results = pd.DataFrame({
    'Modelo': ['Árbol de Decisión', 'Random Forest', 'Gradient Boosting'],

    'RMSE': [
        np.sqrt(mean_squared_error(y_test, y_pred_best_tree)),
        np.sqrt(mean_squared_error(y_test, y_pred_rf)),
        np.sqrt(mean_squared_error(y_test, y_pred_gb))
    ],

    'R2': [
        r2_score(y_test, y_pred_best_tree),
        r2_score(y_test, y_pred_rf),
        r2_score(y_test, y_pred_gb)
    ]
})

results = results.sort_values(by='RMSE')

results
```
